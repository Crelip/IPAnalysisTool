```python
from ip_analysis_tool.util.graph_getter import get_graph_by_date
from ip_analysis_tool.util.date_util import get_date_object
from ip_analysis_tool.k_core import get_max_k_core, k_core_decomposition
from ip_analysis_tool.visualize.graph import visualize_graph, visualize_graph_map
from ip_analysis_tool.util.geo_data_util import get_geo_data
```

```python
# 2013-09-09
week_1_graph = get_graph_by_date(get_date_object('2013-08-05'))
k_core_dec = k_core_decomposition(week_1_graph)
week_1_graph_k_core = get_max_k_core(
    k_core_dec
)

visualize_graph_map(week_1_graph_k_core, name="scratchpad/K-jadro z týždňa začínajúceho 5.8.2013 na mape.png", save=False)
visualize_graph(week_1_graph_k_core, name="scratchpad/K-jadro z týždňa 5.8.2013.png", output_size=(1000, 1000)) # This method does not output visuals directly onto the notebook
print("Number of vertices in the k-core graph:", week_1_graph_k_core.num_vertices())
print("K-value of the k-core graph:", k_core_dec["max_k"])
```

```python
# Details
data : dict = get_geo_data(week_1_graph_k_core.vp.ip[v] for v in week_1_graph_k_core.vertices())
for entry in data.keys():
    print("IP Address:", entry, ", City:", data[entry]["city"]["names"]["en"], "Country:", data[entry]["country"]["names"]["en"], "Province:", data[entry]["subdivisions"][0]["names"]["en"], "Location:", data[entry]["location"])
```

```python
# All time
from ip_analysis_tool.enums import TimeInterval
all_time_graph = get_graph_by_date(time_interval=TimeInterval.ALL)
k_core_dec = k_core_decomposition(all_time_graph)
all_time_graph_k_core = get_max_k_core(
    k_core_dec
)

print("K-core decomposition done")

visualize_graph_map(all_time_graph_k_core, name="scratchpad/K-jadro zo všetkých záznamov na mape.png", save=False)
visualize_graph(all_time_graph_k_core, name="scratchpad/K-jadro zo všetkých záznamov.png", output_size=(2000, 2000))
```
